Lab 23: Def Circle

```python
def circle(radius):
    circumference = 2 * 3.14 * radius
    step = circumference / 360
    for _ in range(360):
        forward(step)
        right(1)
```

Lab 25: Olympics

```python
# 簡化版示意
locs = [(-60, 0), (0, 0), (60, 0), (-30, -30), (30, -30)]
cols = ["blue", "black", "red", "yellow", "green"]
for i in range(5):
    penup(); goto(locs[i]); pendown()
    color(cols[i])
    circle(20) # 假設已有 circle 函式
```

Lab 27: Flower Petal

```python
def draw_petal(size):
    for _ in range(2):
        for _ in range(90): # 1/4 circle
            forward(size)
            right(1)
        right(90)
draw_petal(2)
```

Lab 28: Full Flower

```python
for _ in range(8):
    draw_petal(1)
    right(45)
```
