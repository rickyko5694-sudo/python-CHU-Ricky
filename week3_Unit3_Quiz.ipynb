Q02-01 Odd/Even

```python
n = int(input("請輸入一個整數: "))
result = n % 2
print(f"餘數結果為: {result}")
if n % 2 == 0:
    print(f"{n} 是偶數")
else:
    print(f"{n} 是奇數")
```

Q02-02 F to C

```python
fahrenheit = float(input("請輸入華氏溫度 (°F): "))
celsius = (fahrenheit - 32) * 5 / 9
print(f"華氏 {fahrenheit}°F 等於 攝氏 {celsius:.2f}°C")
```

Q02-03 Exchange

```python
exchange_rate = 32.05
twd_amount = float(input("請輸入台幣金額 (TWD): "))
usd_amount = twd_amount / exchange_rate
print(f"台幣 ${twd_amount:,.0f} 元 約可兌換為 US${usd_amount:,.2f} 元")
```

Q02-04 Time Logic

```python
total_seconds = int(input("請輸入總秒數: "))
hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60
print(f"轉換結果為 - {hours:02d}:{minutes:02d}:{seconds:02d}")
```

Q02-05 Discount

```python
original_price = int(input("請輸入原價: "))
discounted_price = original_price * 0.8
print(f"原價 ${original_price}，打八折後為: ${int(discounted_price)}")
```

Q02-06 Triangle

```python
base = float(input("請輸入三角形的底 (b): "))
height = float(input("請輸入三角形的高 (h): "))
area = (base * height) / 2
print(f"底為 {base}，高為 {height} 的三角形，面積為: {area}")
```

Q02-08 Mutual Exclusive

```python
is_weekend = True
is_holiday = False
if is_weekend or is_holiday:
    print("今天放假！不用上班。")
else:
    print("今天是工作日，加油！")
```

Q02-09 Short-circuit

```python
def check_b():
    print(" B 被執行了！")
    return True
print("開始測試 and...")
result = (1 > 2) and check_b()
print(f"最終結果: {result}\n")
print("開始測試 or...")
result = (2 > 1) or check_b()
print(f"最終結果: {result}")
```

Q02-10 Cashier Bool

```python
total_price = float(input("商品總價: "))
money_held = float(input("你口袋裡的錢: "))
can_afford = money_held >= total_price
print(f"錢夠嗎？ {can_afford}")
if can_afford:
    change = money_held - total_price
    print(f" 付款成功！找零: ${change:,.2f}")
else:
    lack = total_price - money_held
    print(f" 餘額不足，還差 ${lack:,.2f} 元。")
```
