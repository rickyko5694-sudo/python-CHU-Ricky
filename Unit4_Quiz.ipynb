Q03-01 Leap Year

```python
year = int(input("請輸入年份: "))
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} 是閏年")
else:
    print(f"{year} 是平年")
```

Q03-02 Max of 3

```python
a = int(input("輸入第一個數: "))
b = int(input("輸入第二個數: "))
c = int(input("輸入第三個數: "))
result = a
if b > result:
    result = b
if c > result:
    result = c
print(f"最大值是: {result}")
```

Q03-03 Calculator

```python
num1 = float(input("請輸入第一個數字: "))
operator = input("請輸入運算符號 (+, -, *, /): ")
num2 = float(input("請輸入第二個數字: "))
if operator == "+":
    result = num1 + num2
elif operator == "-":
    result = num1 - num2
elif operator == "*":
    result = num1 * num2
elif operator == "/":
    if num2 != 0:
        result = num1 / num2
    else:
        result = "錯誤：除數不可為 0"
else:
    result = "無效的運算符號"
print(f"計算結果: {result}")
```

Q03-04 Order

```python
balance = 100
prices = [120, 80, 250, 60]
if any(balance >= price for price in prices):
    print("還可以吃飯。")
else:
    print("買不起任何一餐。")
```
