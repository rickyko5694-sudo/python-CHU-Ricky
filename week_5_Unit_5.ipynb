P05-01 Create List

```python
fruits = ["Apple","Banana","Orange","Grape","Melon"]
print(fruits)
```

P05-02 Access Items

```python
print(fruits[0]) #Apple
print(fruits[-1]) #Melon
```

P05-03 Update Item

```python
fruits[2] = "Mango"
print(fruits) #Orange 變成了 Mango
```

P05-04 Append

```python
fruits.append("Pear")
```

P05-05 Insert

```python
fruits.insert(1,"Berry") #插在 index 1 的位置
```

P05-06 Remove

```python
if "Banana" in fruits:
  fruits.remove("Banana")
```

P05-07 Pop

```python
last_fruit = fruits.pop()
print(f"I ate {last_fruit}")
```

P05-08 Length

```python
print(len(fruits))
```

P05-09 Sort

```python
nums = [3, 1, 4, 1, 5]
nums.sort()
print(nums) #[1, 1, 3, 4, 5]
```

P05-10 Reverse

```python
nums.reverse()
```

```python
print(fruits[1:4])
```

```python
for f in fruits:
    print(f)
```

P05-14 Check In

```python
if "Apple" in fruits:
  print("Has Apple")
```

```python
print(fruits.index("Apple"))
```

P05-16 Count

```python
print(fruits.count("Apple"))
```
