P03-01 Simple If

```python
x = 10
if x > 0:
  print("Positive")
```

P03-02 If-Else

```python
x = -5
if x > 0:
  print("Pos")
else:
  print("Non-Pos")
```

P03-03 Odd/Even

```python
n = 4
if n % == 0:
  print("Even")
else:
  print("Odd")
```

P03-04 Pass Check

```python
score = 59
if score >= 60:
  print("Pass")
else:
  print("Fail")
```

P03-05 Password

```python
pwd = input("Pwd: ")
if pwd == "secret":
  print("Access Granted")
else:
  print("Denied")
```

P03-06 Elif

```python
n = 0
if n > 0: print("+")
else n < 0: print("-")
```

P03-07 Grades

```python
s = 85
if s >= 90: print("A")
elif s >= 80: print("B") #此時 s 一定 < 90
else: print("C")
```

P03-08 Nested If

```python
age = 20
licence = True
if age >= 18:
  if licence:
    print("Can Drive")
```

P03-09 Age Group

```python
age = 5
if age < 12: print("Child")
elif age < 18: print("Teen")
else: print("Adult")
```

P03-10 Multiple Cond
